# Genres movies

```python
url = 'https://api.themoviedb.org/3/genre/movie/list?api_key=021e610e76806462bccde348f1756293&language=en-US'

response = requests.get(url)
data = response.json()
genres_movies = pd.DataFrame(data)


def flatten(genres_movies, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(genres_movies[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        genres_movies = pd.concat([genres_movies, flattened], axis=1)
        genres_movies = genres_movies.drop(column, axis=1)
    return genres_movies

nested_columns = ['genres']

genres_movies_flat = flatten(genres_movies, nested_columns)
genres_movies_flat
```

# Movies 2019

```python
# Start with an empty list
total_results = []

# Loop through from pages 1 to 500
for page_num in range(1, 501):
    # Build the URL and download the results
    url = "https://api.themoviedb.org/3/discover/movie?api_key=021e610e76806462bccde348f1756293&primary_release_year=2019"
    response = requests.get(url)
    data = response.json()
    total_results = total_results + data['results']


print("We have", len(total_results), "total results")
```

```python
# loading the data to a dataframe

movies2019 = pd.DataFrame(total_results)
movies2019.head()
```

# Final database

```python
# define column names for our new dataframe
columns = ['Title', 'Movie_ID', 'Imdb_ID', 'Language', 'Release Date', 'Genre', 'Popularity', 'Vote average', 'Vote count','Revenue USD', 'Budget USD']

# create dataframe 
movies_df = pd.DataFrame(columns=columns)

# making a request for the API
for movie in total_results:
    movie_data = requests.get('https://api.themoviedb.org/3/movie/'+ str(movie['id']) +'?api_key='+ '021e610e76806462bccde348f1756293')
    movie_data = movie_data.json()
    movies_df.loc[len(movies_df)]=[movie['title'], movie_data['id'], movie_data['imdb_id'], movie_data['original_language'], movie_data['release_date'], movie_data['genres'], movie_data['popularity'], movie_data['vote_average'], movie_data['vote_count'], movie_data['revenue'], movie_data['budget']] 
```

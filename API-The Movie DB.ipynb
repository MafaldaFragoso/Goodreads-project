# Genres movies

```python
url = 'https://api.themoviedb.org/3/genre/movie/list?api_key=xxx&language=en-US'

response = requests.get(url)
data = response.json()
genres_movies = pd.DataFrame(data)


def flatten(genres_movies, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(genres_movies[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        genres_movies = pd.concat([genres_movies, flattened], axis=1)
        genres_movies = genres_movies.drop(column, axis=1)
    return genres_movies

nested_columns = ['genres']

genres_movies_flat = flatten(genres_movies, nested_columns)
genres_movies_flat
```

# Movies 2019

```python
# Start with an empty list
total_results = []


# Loop through from pages 1 to 100
for page_num in range(1, 100):
    # Build the URL and download the results
    url = "https://api.themoviedb.org/3/discover/movie?api_key=xxx&sort_by=popularity.desc&page="+ str(page_num)+"&primary_release_year=2019"
    response = requests.get(url)
    data = response.json()
    total_results = total_results + data['results']

    
print("We have", len(total_results), "total results")
```

# Simulation

```python
df_id_revenue = pd.DataFrame(columns = ['id', 'imdb_id', 'revenue', 'budget'])

for id_number in list_ids:
    response = requests.get('https://api.themoviedb.org/3/movie/'+ str(id_number) +'?api_key='+ 'xxx')
    movie_data = response.json()
    if response.status_code == 200:
        df_id_revenue = df_id_revenue.append({'id': movie_data['id'], 
                                          'imdb_id': movie_data['imdb_id'], 
                                          'revenue': movie_data['revenue'], 
                                          'budget': movie_data['budget']}, 
                                         ignore_index=True)  
    else:
        print("Movie ", id_number, " does not have details")
```

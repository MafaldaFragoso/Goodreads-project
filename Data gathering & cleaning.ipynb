# Data cleaning

```python
# importing the file

data_project = pd.read_csv('data_movies_2019.csv')
data_project.head()
```

```python
# checking the shape of the dataframe

data_project.shape
```

```python
# checking the names of the columns of the dataframe
print(list(data_project.columns))
```

```python
# checking information about the dataframe

data_project.info()
```

```python
# changing the datatype of the column id to object

data_project['id'] = data_project['id'].astype('object')
data_project['id'].dtype
```

```python
# changing the datatype of the column runtimeMinutes to float

data_project['runtimeMinutes'] = pd.to_numeric(data_project['runtimeMinutes'],errors='coerce')
data_project['runtimeMinutes'].dtype
```

```python
# checking for missing values

null_cols = data_project.isnull().sum()
null_cols[null_cols > 0]
```

```python
# checking for low variance

low_variance = []

for col in data_project._get_numeric_data():
    minimum = min(data_project[col])
    ninety_perc = np.percentile(data_project[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)

print(low_variance)
```

```python
# creating a new column for the boolean isAdult column with the values of 0: not adult only; 1: adult only

data_project.loc[data_project['isAdult'] == 0, 'Audience'] = 'not adult only'
data_project.loc[data_project['isAdult'] == 1, 'Audience'] = 'adult only'
```

```python
# dropping the isAdult column

data_project = data_project.drop(['isAdult'], axis=1)
```

```python
data_project.head()
```

```python
# statistics

stats = data_project.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
# filtering data with records of revenue higher than zero

filtered_revenue = data_project[(data_project['revenue']>0)]

stats = filtered_revenue.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
# filtering data with records of budget higher than zero

filtered_budget = data_project[(data_project['budget']>0)]

stats = filtered_budget.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
# checking the minimum and maximum values in the column runtimeMinutes

print("Max duration:", data_project['runtimeMinutes'].max())
print("Min duration:", data_project['runtimeMinutes'].min())
```

```python
# filtering data with records of runtimeMinutes below 20

filtered_duration = data_project[(data_project['runtimeMinutes']<20)]
filtered_duration.head()

# we can see that the movies with lower duration time are classified as a short movie
```

```python
# checking the minimum and maximum values in the column numVotes

print("Max number of votes: ", data_project['numVotes'].max())
print("Min number of votes: ", data_project['numVotes'].min())
```

```python
# checking for outliers by plotting the boxplot for each variable
# for the variables revenue and bidget I've used the filtered_revenue and filtered_budget dataframe 

fig, ax = plt.subplots(1,5, figsize=(16, 4))
sns.boxplot(y=filtered_revenue.revenue, ax=ax[0])
sns.boxplot(y=filtered_budget.budget, ax=ax[1])
sns.boxplot(y=data_project.runtimeMinutes, ax=ax[2])
sns.boxplot(y=data_project.averageRating, ax=ax[3])
sns.boxplot(y=data_project.numVotes, ax=ax[4])
plt.subplots_adjust(wspace=0.5)
```

```python
genres = data_project["genres"].str.split("," , expand = True)

data_project["Genre 1"]= genres[0]
data_project["Genre 2"]= genres[1]
data_project["Genre 3"]= genres[2]
```

```python
data_project.head()
```

```python
unique_genres = ['Action', 'Adult', 'Adventure', 'Animation', 'Biography','Comedy', 'Crime', 'Documentary', 
                 'Drama', 'Family', 'Fantasy', 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'Romance',
                 'Sci-Fi', 'Sport', 'Talk-Show', 'Thriller', 'War', 'Western', 'Reality-TV']
```

```python
for genre in unique_genres:
    data_project[genre] = 0
data_project
```

```python
for i, row in data_project.iterrows():
    for genre in unique_genres:
        if (genre == row['Genre 1'] or genre == row['Genre 2'] or genre == row['Genre 3']):
            data_project.at[i, genre] = 1
        
```

```python
# dividing the revenue and budget title for one million to better scale with other variables

data_project['revenue'] = data_project['revenue']/1000000
data_project['budget'] = data_project['budget']/1000000
```

```python
# renaming the columns revenue and budget
data_project = data_project.rename(columns={'revenue':'revenue_USDm',
                                            'budget':'budget_USDm'})
```

```python
# dropping the columns primary Title, end year and genres

data_project = data_project.drop(['primaryTitle'], axis=1)
data_project = data_project.drop(['endYear'], axis=1)
data_project = data_project.drop(['genres'], axis=1)
```

```python
# printing the final dataset with only the records for the year 2019 

data_final = data_project[(data_project['startYear']=='2019')]
data_final.head()
```

```python
data_final.shape
```

```python
# printing the final dataset with only the records for runtimeMinutes higher than 0

data_final2 = data_final[(data_final['runtimeMinutes']>0)]
data_final2.head()
```

```python
data_final2.shape
```

```python
#checking again for duplicate values

before = len(data_final2)
data_final3 = data_final2.drop_duplicates()
after = len(data_final3)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
# exporting to a csv file
data_final2.to_csv('data_clean2.csv', index=False)
```

```python
# filtering the data to only the records with revenue and budget

filtered_financials = data_final[(data_final['revenue_USDm']>0) & (data_final['budget_USDm']>0)]
filtered_financials
```

```python
# exporting to a csv file
filtered_financials.to_csv('data_clean_financials.csv', index=False)
```
